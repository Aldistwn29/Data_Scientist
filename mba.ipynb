```python
import numpy as np
import pandas as pd
data_transaksi = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/transaksi_dqlab_retail.tsv", sep="\t")
```

```python
data_transaksi
```

```python
data_transaksi.info()
```

```python
# mengelompokan data pada kranjang yg sama
data_transaksi['Flag'] = 1
```

```python
# melakukan pengelompokan flag berdasarkan kolom kode transaksi dan nama barang
basket = data_transaksi.groupby(['Kode Transaksi', 'Nama Barang'])['Flag'].sum().unstack().reset_index().fillna(0).set_index('Kode Transaksi')
basket.head()
```

```python
# membuat function untuk menormalisasikan data
def encode_units(x):
  if x <= 0:
    return 0
  if x > 0:
    return 1

# menerapkan fungsi encode
basket_encode = basket.applymap(encode_units)
basket_encode
```

```python
# mengimport algoritma apriori
from mlxtend.frequent_patterns import apriori
frequent_itemset = apriori(basket_encode, min_support=0.03, use_colnames=True)
frequent_itemset
```

```python
#Mengimport association_rules
from mlxtend.frequent_patterns import association_rules

#Menerapkan association_rules berdasarkan frequent_itemset
rules =  association_rules(frequent_itemset, metric='lift', min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
rules
```
